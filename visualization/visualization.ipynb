## Visualization code for Scores distribution and relation_id

```python
import numpy as np
import json
import pandas as pd
import seaborn
```

```python
records = []
with open('knowns_return_mlp.json', "r") as f:
    for line in f:
        records.append(json.loads(line))
```

```python
len(records)
```

```python
records[0]
```

```python
df = pd.DataFrame.from_records(records, index=['known_id'])
```

```python
df['edit_layer'] = df.apply(lambda row: np.argmax(row.scores[row.subject_range[1]]), axis=1)
```

```python
seaborn.set(rc={'figure.figsize':(15,10)})
seaborn.scatterplot(data=df[['relation_id', 'edit_layer']], x='relation_id',y='edit_layer', hue='relation_id')
```

```python
seaborn.set(rc={'figure.figsize':(10,6)})
seaborn.kdeplot(df[['relation_id', 'edit_layer']])
```
